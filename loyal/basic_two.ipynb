# Encoding of Categorial Column

```python
dt={'Loyal Customer' : 1,
                    'disloyal Customer' : 0}
#dt.keys()
#dt.values()
df['Customer Type']=df['Customer Type'].map(dt)
```

```python
df['Customer Type'].unique()
```

```python
dt1={'Male' : 1,
                    'Female' : 0}               #dictionary
#dt.keys()
#dt.values()
df['Gender']=df['Gender'].map(dt1)              #mapping of dictionary in feature
```

```python
df['Gender'].unique()
```

```python
dt2={'Personal Travel' : 1,
                    'Business travel' : 0}                 #dictionary

df['Type of Travel']=df['Type of Travel'].map(dt2)           #mapping of dictionary in feature
```

```python
df['Type of Travel'].unique()
```

```python
dt3={'Eco Plus' : 1,
    'Business' : 0,
    'Eco' : 2
    }                     #dictionary

df['Class']=df['Class'].map(dt3)                     #mapping of dictionary in feature
df['Class'].unique()
```

```python
dt4={'neutral or dissatisfied' : 1,
                    'satisfied' : 0}               #Dictionary

df['satisfaction']=df['satisfaction'].map(dt4)                #Mapping of dictionary in feature
df['satisfaction'].unique()
```

# Split Data

```python
from sklearn.model_selection import train_test_split
```

```python
x=df.drop(['Customer Type'],axis=1)
```

```python
x
```

```python
y= df['Customer Type']
y
```

```python
df.duplicated().sum()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

# Train a Model

```python
lr.fit(x_train,y_train)
```

```python
print("Training Score:",lr.score(x_train,y_train))
print("Training Score:",lr.score(x_test,y_test))
```

```python
pred=lr.predict(x_test)
pred
```
